```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
```

```python
# Step 1: Create Sample Dataset
data = pd.DataFrame({
    'Math': [85, 70, 60, 40, 30],
    'Science': [90, 75, 65, 50, 35],
    'English': [88, 78, 68, 55, 38]
})
```

```python
# Step 2: Standardize the Data
scaler = StandardScaler()
data_std = scaler.fit_transform(data)
```

```python
# Step 3: Apply t-SNE
tsne = TSNE(n_components=2, perplexity=3, random_state=42)
tsne_results = tsne.fit_transform(data_std)
```

```python
# Step 4: Convert to DataFrame
tsne_df = pd.DataFrame(tsne_results, columns=['TSNE1', 'TSNE2'])
tsne_df['Student'] = ['A', 'B', 'C', 'D', 'E']
```

```python
# Step 5: Plot the Results
plt.figure(figsize=(8, 5))
plt.scatter(tsne_df['TSNE1'], tsne_df['TSNE2'], c=['red', 'blue', 'green', 'purple', 'orange'])
for i, txt in enumerate(tsne_df['Student']):
    plt.annotate(txt, (tsne_df['TSNE1'][i], tsne_df['TSNE2'][i]), fontsize=12)
plt.title("t-SNE Visualization of Student Scores")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.show()
```
